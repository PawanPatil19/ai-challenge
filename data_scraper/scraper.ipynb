```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By

from time import sleep
from pprint import pprint
```

```python
def get_search_results(num_pages = 25):
    driver = webdriver.Chrome()
    driver.set_window_size(1024, 600)
    driver.maximize_window()
    base = 'https://noteb.com/?'
    page_links = []
    total_count = 0
    for page in range(1, num_pages + 1):
        print(page)
        url = base + f'search/search.php?advsearch=1&Regions[]=United%20States&bdgminadv=120&bdgmaxadv=9600&exchadv=USD&nrcoresmin=2&nrcoresmax=24&cputdpmin=3&cputdpmax=157&cpufreqmin=1.1&cpufreqmax=5.8&cputechmin=2&cputechmax=28&launchdatemin=2022&launchdatemax=2024&gputype=2&gpumemmin=2048&gpumemmax=32768&gpubusmin=64&gpubusmax=256&gpulaunchdatemin=2015&gpulaunchdatemax=2024&gpupowermin=10&gpupowermax=250&displaymin=10.1&displaymax=18&verresmin=768&verresmax=2400&capacitymin=16&capacitymax=65536&nrhdd=1&mdbslots=0&mdbwwan=0&rammin=4&rammax=192&freqmin=1600&freqmax=7500&batlifemin=1.4&batlifemax=19.7&acumcapmin=24&acumcapmax=100&weightmin=0.5&weightmax=966.5&thicmin=7.4&thicmax=356&widthmin=19.9&widthmax=415.5&depthmin=21.3&depthmax=330&webmin=0&webmax=9&wnetspeed=150&oddtype=Any/None&yearsmin=0&yearsmax=3&page={page}&sort_by=value'
        driver.get(url)
        WebDriverWait(driver, 10)
        # Necessary to wait for the page to load
        sleep(5)
        links = driver.find_elements(By.CLASS_NAME,'searchresult-content')
        links = list(map(lambda x: (x.get_dom_attribute('onmousedown').split("'")[1], x.text), links))
        total_count += len(links)
        print("Laptops: ", total_count)
        page_links.extend(links)
    driver.quit()
    return page_links
    
search_results = get_search_results()
print(search_results)
```

```python
# json_keys = [
#     "Name", "Screen Size", "Processor", "Graphics", "RAM", "Storage",
#     "Weight", "OS", "Battery Life", "Rating", "Price", "Touchscreen"
# ]
from selenium.webdriver.common.by import By
import json
from concurrent.futures import ThreadPoolExecutor

def parse_header(txt):
    """
    Chuwi GemiBook Pro
    14.0" - 2160x1440
    INTEL N100
    INTEL UHD Graphics
    8 GB DDR4
    256 GB - SSD
    1.5 Kg / 3.31 lb
    Windows 11 Home
    Estimated battery:
    7.6 - 8.1 h
    """
    name, screen, processor, graphics, ram, storage, weight, os, _, battery = txt.split('\n')
    screen_size, screen_res = screen.split(' - ')
    return {
        "name": name,
        "screen_size": screen_size,
        "screen_resolution": screen_res,
        "processor": processor,
        "graphics": graphics,
        "ram": ram,
        "storage": storage,
        "weight": weight,
        "os": os,
        "battery_life": battery,
    }
    
    
def scrape_info(link, info):
    """
    Scrapes the additional information from the laptop detail page.
    """
    options = webdriver.ChromeOptions()
    options.add_argument('--headless')  # To run Chrome in headless mode (without opening browser window)
    driver = webdriver.Chrome(options=options)
    url = 'https://noteb.com/' + link
    driver.get(url)
    WebDriverWait(driver, 10)
    sleep(3)
    config_price1 = driver.find_element(By.ID, 'config_price1')
    config_price2 = driver.find_element(By.ID, 'config_price2')
    price = config_price1.text + '-' + config_price2.text
    rating = driver.find_element(By.CLASS_NAME, 'progressBar__percent').text
    touchscreen = driver.find_element(By.XPATH, '/html/body/div[3]/div/div[2]/div[1]/div/div[4]/div[1]/div[2]/div[2]/div[2]/div/div/div[7]/div[2]').text
    dct = parse_header(info)
    dct["price"] = price
    dct["rating"] = rating
    dct["touchscreen"] = touchscreen
    dct["url"] = url
    print(dct["name"])
    driver.quit()
    return dct

def get_info(search_results):
    """
    Gets laptop information from search results.
    """
    base = 'https://noteb.com/?'
    arr = []
    with ThreadPoolExecutor(max_workers=5) as executor:  # Adjust max_workers as needed
        futures = []
        for link, info in search_results:
            futures.append(executor.submit(scrape_info, link, info))
        for future in futures:
            arr.append(future.result())
            with open('noteb.json', 'w') as f:
                json.dump(arr, f, indent=4)
    return arr

info = get_info(search_results)
    
```
